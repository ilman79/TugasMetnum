```python
import numpy as np
import math
```

```python
#untuk methode bisection
def bisection(f, a, b, e=10**-5, N = 100):
  x_a = a
  x_b = b
  
  if f(x_a)*f(x_b) >= 0:
    print("Use another guess range")
    return None
  
  for n in range (1,N):
    x_m = (x_a + x_b)/2
    print('Iterasi = ',abs(f(x_m)))
    if abs(f(x_m)) < e:
      return (x_m)
    else:
      if f(x_a)*f(x_m) < 0:
        x_b = x_m
      elif f(x_b)*f(x_m) < 0:
        x_a = x_m
  return (x_m)

f = lambda x: (np.e)**x + 2**-x + 2*math.cos(x) - 6
a = 1
b = 2
x_root = bisection(f,a,b)

print(f'\n akar dari f(x) : {x_root}')
```

```python
#untuk methode Fixed point
from numpy.linalg import norm
def fixedp(f,x0,tol=10e-5,maxiter=100):
 """ Fixed point algorithm """
 e = 1
 itr = 0
 xp = []
 while(e > tol and itr < maxiter):
  x = f(x0)      # fixed point equation
  e = norm(x0-x) # error at the current step
  x0 = x
  xp.append(x0)  # save the solution of the current step
  itr = itr + 1
 return x,xp
```

```python
f = lambda x : math.sqrt(x)

x_start = .5
xf,xp = fixedp(f,x_start)
```

```python
#untuk methode Newton
def newt(h,dh,x0):
  e = 10**-4
  N = 100
  for i in range (0,N):
    print("%d   | %f    | %f    " %(i,x0,f(x0)))
    if abs(h(x0)) < e:
      return x0,i
    if dh(x0) == 0:
      print("0")
      return None
    x0 = x0 - h(x0)/dh(x0)
  print("Maksimal perulangan")
  return x0, i

h  = lambda x: (np.e)**x + 2**-x + 2*np.cos(x)-6
dh = lambda x: (np.e)**x - 2**(-x-1) - 2*np.sin(x)
x0 = 1

x_akar, iterasi= newt(h,dh,x0)
print('Hasil : ', x_akar)
print('In %d iterasi' %iterasi)
```

```python
def secant(f,x0,x1,e = 10**-5,N=100):
  for i in range (0,N):
    print("%d   | %f    | %f    " %(i,x1,f(x1)))
    if abs(f(x1)) < e:
      return x1,i
    x2 = x1 - f(x1)*(x1-x0)/(f(x1)-f(x0))
    x0 = x1
    x1 = x2
  print("Maximum iteration")
  return x1, i

f  = lambda x: (np.e)**x + 2**-x + 2*np.cos(x)-6
x0 = 1
x1 = 2

x_root, iteration = secant(f,x0,x1)
print('Result : ', x_root)
print('In %d iteration' %iteration)
```
